### Mock Charts - multi-node, perception vs non perception, eCloudSim

```python
# Fake data for the 4 scenarios
data_2p = np.random.normal(loc=200, scale=20, size=100)
data_4p = np.random.normal(loc=250, scale=20, size=100)
data_8p = np.random.normal(loc=300, scale=20, size=100)
data_16p = np.random.normal(loc=350, scale=20, size=100)
data_32p = np.random.normal(loc=500, scale=20, size=100)

data_2np = np.random.normal(loc=50, scale=10, size=100)
data_4np = np.random.normal(loc=100, scale=10, size=100)
data_8np = np.random.normal(loc=150, scale=10, size=100)
data_16np = np.random.normal(loc=250, scale=10, size=100)
data_32np = np.random.normal(loc=450, scale=10, size=100)
data_64np = np.random.normal(loc=550, scale=10, size=100)
data_128np = np.random.normal(loc=600, scale=10, size=100)

sns.set(font_scale=1.2)
sns.set_style('whitegrid')
# Combine the data into single numpy arrays
data_p_total = np.concatenate([data_2p, data_4p, data_8p, data_16p, data_32p])
data_np_total = np.concatenate([data_2np, data_4np, data_8np, data_16np, data_32np, data_64np, data_128np])

# Create corresponding labels for each data array
labels_p = ['2 Cars']*100 + ['4 Cars']*100 + ['8 Cars']*100 + ['16 Cars']*100 + ['32 Cars']*100
labels_np = ['2 Cars']*100 + ['4 Cars']*100 + ['8 Cars']*100 + ['16 Cars']*100 + ['32 Cars']*100 + ['64 Cars']*100 + ['128 Cars']*100

# Convert the numpy array and labels to pandas DataFrames
df_p_total = pd.DataFrame({'Number of Cars': labels_p, 'Total Simulation Time (ms)': data_p_total})
df_np_total = pd.DataFrame({'Number of Cars': labels_np, 'Total Simulation Time (ms)': data_np_total})

# Create box plots for total simulation time
plt.figure(figsize=(10, 6))
ax1 = sns.boxplot(data=df_p_total, x='Number of Cars', y='Total Simulation Time (ms)')
ax1.set_xlabel('Number of Vehicles')
ax1.set_ylabel('Total Simulation Time (ms)')
ax1.set_title('eCloudSim: Total Simulation Time\nper Number of Vehicles (with perception)')
ax1.set_xticklabels([2, 4, 8, 16, 32])
plt.show()

plt.figure(figsize=(10, 6))
ax2 = sns.boxplot(data=df_np_total, x='Number of Cars', y='Total Simulation Time (ms)')
ax2.set_xlabel('Number of Vehicles')
ax2.set_ylabel('Total Simulation Time (ms)')
ax2.set_title('eCloudSim: Total Simulation Time\nper Number of Vehicles (without perception)')
ax2.set_xticklabels([2, 4, 8, 16, 32, 64, 128])
plt.show()

# Fake data for the 4 scenarios
data_2p_step = np.random.normal(loc=5, scale=1, size=100)
data_4p_step = np.random.normal(loc=6, scale=1, size=100)
data_8p_step = np.random.normal(loc=7, scale=1, size=100)
data_16p_step = np.random.normal(loc=8, scale=1, size=100)
data_32p_step = np.random.normal(loc=10, scale=1, size=100)

data_2np_step = np.random.normal(loc=5, scale=1, size=100)
data_4np_step = np.random.normal(loc=6, scale=1, size=100)
data_8np_step = np.random.normal(loc=7, scale=1, size=100)
data_16np_step = np.random.normal(loc=8, scale=1, size=100)
data_32np_step = np.random.normal(loc=10, scale=1, size=100)
data_64np_step = np.random.normal(loc=12, scale=1, size=100)
data_128np_step = np.random.normal(loc=14, scale=1, size=100)


# Create box plots for simulation step time
df_p_step = pd.DataFrame({'Number of Cars': labels_p, 'Simulation Step Time (ms)': data_p_total})
df_np_step = pd.DataFrame({'Number of Cars': labels_np, 'Simulation Step Time (ms)': data_np_total})

# Create box plots for simulation step time
plt.figure(figsize=(10, 6))
ax3 = sns.boxplot(data=df_p_step, x='Number of Cars', y='Simulation Step Time (ms)')
ax3.set_xlabel('Number of Vehicles')
ax3.set_ylabel('Simulation Step Time (ms)')
ax3.set_title('eCloudSim: Simulation Step Time\nper Number of Vehicles (with perception)')
ax3.set_xticklabels([2, 4, 8, 16, 32])
plt.show()

plt.figure(figsize=(10, 6))
ax4 = sns.boxplot(data=df_np_step, x='Number of Cars', y='Simulation Step Time (ms)')
ax4.set_xlabel('Number of Vehicles')
ax4.set_ylabel('Simulation Step Time (ms)')
ax4.set_title('eCloudSim: Simulation Step Time\nper Number of Vehicles (without perception)')
ax4.set_xticklabels([2, 4, 8, 16, 32, 64, 128])
plt.show()
```

### Mock Chart - GPU Utilization, Perception, Single node

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fake data for OpenCDA and eCloudSim (with perception on single node)
num_vehicles = [2, 4, 8, 16, 32]
opencda_gpu_util = [50, 57, 100, 100, 100]
ecloud_gpu_util = [49, 53, 100, 100, 100]

# Create a pandas DataFrame
df_gpu_util = pd.DataFrame({'Number of Vehicles': num_vehicles * 2,
                            'GPU Memory Utilization (%)': np.concatenate([opencda_gpu_util, ecloud_gpu_util]),
                            'Simulator': ['OpenCDA'] * len(num_vehicles) + ['eCloudSim'] * len(num_vehicles)})

# Create a line plot using seaborn
plt.figure()
ax = sns.lineplot(data=df_gpu_util, x='Number of Vehicles', y='GPU Memory Utilization (%)', hue='Simulator', style='Simulator', markers=True)
plt.xticks(num_vehicles)

# Add labels and titles
ax.set_xlabel('Number of Vehicles')
ax.set_ylabel('Avg. GPU Memory Utilization (%)')
ax.set_title('GPU Memory Utilization vs Number of Vehicles\n(Perception, Single Node)')

# Display the plot
plt.show()
```
