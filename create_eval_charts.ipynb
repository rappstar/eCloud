```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Data
vehicles = [2, 4, 8, 16, 32]
eCloudSim = [210, 250, 280, 320, 423]
openCDA = [230, 270, 300, 360, 450]
apollo = [289, 300, 350, 420, 460]

# Plotting
sns.set_style("whitegrid")
sns.set_palette("deep")
sns.set_context("talk")

plt.plot(vehicles, eCloudSim, marker='o')
plt.plot(vehicles, openCDA, marker='o')
plt.plot(vehicles, apollo, marker='o')

plt.xlabel('Number of Vehicles')
plt.xticks(vehicles)
plt.ylabel('Total Simulation Time (s)')
plt.title('Simulation Time without Perception on a Single Node\n(Multi2lane Scenario)')

plt.legend(['eCloudSim', 'Original OpenCDA', 'Apollo'])

plt.ylim([200, 500])

plt.show()
```

```python
sns.set_style('whitegrid')

# Fake data
data_4 = np.random.normal(loc=5, scale=1, size=100)
data_8 = np.random.normal(loc=7, scale=1, size=100)
data_16 = np.random.normal(loc=10, scale=2, size=100)

# Combine the data into a single numpy array
data = np.concatenate([data_4, data_8, data_16])

# Create corresponding labels for each data array
labels = ['4 Cars']*100 + ['8 Cars']*100 + ['16 Cars']*100

# Convert the numpy array and labels to a pandas DataFrame
df = pd.DataFrame({'Number of Cars': labels, 'Simulation Step Time (ms)': data})

# Create a box plot using seaborn
ax = sns.boxplot(data=df, x='Number of Cars', y='Simulation Step Time (ms)')

# Add labels and titles
ax.set_xlabel('Number of Cars')
ax.set_ylabel('Simulation Step Time (ms)')
ax.set_title('Simulation Step Time per Number of Cars')

# Display the plot
plt.show()
```
