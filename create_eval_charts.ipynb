```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#fake data
data = {'OpenCDA with Perception': 360,
        'eCloud with Perception': 250,
        'OpenCDA without Perception': 180,
        'eCloud without Perception': 120}

sns.barplot(x = list(data.keys()), y = list(data.values()))
plt.title("Total Simulation Time\n (8 cars - Multi2lane Scenario)")
plt.ylabel("Average Simulation Time (seconds)")
plt.xticks(rotation = 30)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_style('whitegrid')

# Fake data
data = {'OpenCDA with Perception': [360, 420, 500],
        'eCloud with Perception': [250, 300, 350],
        'OpenCDA without Perception': [180, 240, 300],
        'eCloud without Perception': [120, 180, 240]}

# Create a list of x values that represent the number of cars
x_values = [4, 8, 16]

# Convert the dictionary of data to a pandas DataFrame
df = pd.DataFrame(data, index=x_values)

# Create a line plot using seaborn
for column in df.columns:
    sns.lineplot(x=x_values, y=df[column], label=column, markers=True, dashes=False)
    
# Add labels and titles
plt.title("Total Simulation Time\n (Multi2lane Scenario)")
plt.xlabel("Number of Cars")
plt.ylabel("Average Simulation Time (seconds)")

# Add a legend with appropriate labels
plt.legend(title='Simulation Type')

# Set the y axis limits based on the data range
plt.ylim([0, np.max(list(data.values())) + 50])

plt.show()
```

```python
sns.set_style('whitegrid')

# Fake data
data_4 = np.random.normal(loc=5, scale=1, size=100)
data_8 = np.random.normal(loc=7, scale=1, size=100)
data_16 = np.random.normal(loc=10, scale=2, size=100)

# Combine the data into a single numpy array
data = np.concatenate([data_4, data_8, data_16])

# Create corresponding labels for each data array
labels = ['4 Cars']*100 + ['8 Cars']*100 + ['16 Cars']*100

# Convert the numpy array and labels to a pandas DataFrame
df = pd.DataFrame({'Number of Cars': labels, 'Simulation Step Time (ms)': data})

# Create a box plot using seaborn
ax = sns.boxplot(data=df, x='Number of Cars', y='Simulation Step Time (ms)')

# Add labels and titles
ax.set_xlabel('Number of Cars')
ax.set_ylabel('Simulation Step Time (ms)')
ax.set_title('Simulation Step Time per Number of Cars')

# Display the plot
plt.show()
```
